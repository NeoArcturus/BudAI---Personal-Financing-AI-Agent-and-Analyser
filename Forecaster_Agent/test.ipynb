```python
import pandas as pd
import matplotlib.pyplot as mp
```

```python
df = pd.read_csv("all_paths.csv", header=None)
df.head(20)
```

```python
mp.figure(figsize=(20, 10))
        
for i in range(len(df)):
    mp.plot(df.iloc[i], color='blue')
    
mean_path = df.mean(axis=0)

mp.title('Monte Carlo GBM Forecast (60 Days)')
mp.xlabel('Days into the Future')
mp.ylabel('Projected Balance (£)')
mp.grid(True, alpha=0.3)
mp.legend()
mp.tight_layout()
```
